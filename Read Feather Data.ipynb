```python
#Import necessary libraries
import pandas as pd

#Read data
data = pd.read_feather(r"C:\Users\omerf\OneDrive\Masaüstü\Data Analysis Project\Sales Data/Sales_data.ftr")
data.head()
```

```python
#Check for missing values
data.isnull().sum()
```

```python
#Check for duplicates
data[data.duplicated()]
```
